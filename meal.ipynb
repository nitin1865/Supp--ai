```python
import streamlit as st
import pandas as pd
import os
from langchain.agents import Tool, AgentExecutor, create_react_agent
from langchain_community.tools import DuckDuckGoSearchRun
# import google.generativeai as genai
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
import matplotlib.pyplot as plt
import re

st.set_page_config(page_title="🥗 Personalized Meal Plan Generator", layout="wide")

# ---------- CONFIGURE API ----------
# api_key = "AIzaSyBMdmH6lJdWLtqQ4YBB8bkqXaGukQm_47s"
# genai.configure(api_key=api_key)
# model = genai.GenerativeModel('gemini-2.0-flash-001')

# Initialize the Search Tool
search = DuckDuckGoSearchRun()

tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Useful for searching information on the internet. Use this when you need to find current or factual information."
    )
]

# Initialize the language model
llm = ChatGroq(
    temperature=0.7,
    model="llama3-8b-8192",
    api_key=os.getenv("GROQ_API_KEY"),
)
# ---------- LOAD MEAL CSV ----------
@st.cache_data
def load_meals(csv_path: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])
    return df

meals_df = load_meals("meal.csv")

# ---------- BMI and BMR functions ----------
def classify_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)

def calculate_bmr(weight_kg, height_cm, age, gender):
    if gender.lower().startswith('m'):
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age + 5, 1)
    else:
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age - 161, 1)

# ---------- MEAL RECOMMENDATION ----------
def meal_recommendation(user_profile: dict, meals_df: pd.DataFrame) -> str:
    bmi_value = classify_bmi(user_profile['weight'], user_profile['height'])
    bmr = calculate_bmr(
        user_profile['weight'], user_profile['height'], user_profile['age'], user_profile['gender']
    )
    prompt = f"""
You are an expert dietitian assistant.

User Profile:
- Name: {user_profile['name']}
- Age: {user_profile['age']}
- Gender: {user_profile['gender']}
- Height: {user_profile['height']} cm
- Weight: {user_profile['weight']} kg
- BMI: {bmi_value}
- BMR: {bmr} kcal/day
- Health Goals: {', '.join(user_profile['health_goals'])}
- Food Allergies: {', '.join(user_profile['food_allergies'])}
- Dietary Pattern: {user_profile['diet']}
- Disease: {', '.join(user_profile['disease'])}
- Activity Level: {user_profile['activity_level']}
- Supplement Preferences: {', '.join(user_profile['supplement_preferences'])}

---

Task:
- Create a personalized meal plan with exactly 5 meals per day: Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, and Dinner.
- Choose **only** from this list of meals given below. Do not invent new meals.
- Each meal in the list has type (Breakfast, Snack, Lunch, Dinner), calories, protein, carbs, and fats.
- Consider user's BMI:
  * If BMI <18.5: Recommend meals above BMR (weight gain).
  * If BMI >=25: Recommend meals below BMR (weight loss).
  * If BMI 18.5–25: Recommend around BMR (maintenance).

---

📄 **Available Meals**:

{meals_df.to_string(index=False)}

---

✅ Format your answer like this:

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only pick from the provided list. Keep it personalized and explain briefly why you chose each.
"""
    # response = model.generate_content(prompt)
    response = llm.invoke(prompt)
    # return response.text.strip()
    return response.text

# ---------- MEAL CONFIGURATION BY USER ----------
def meal_change_according_to_user(user_input: str, recommended_meal: str, meals_df: pd.DataFrame) -> str:
    prompt = f"""
You are an expert dietitian assistant.

Below is the meal plan previously recommended to the user:

{recommended_meal}

The user has requested the following change:
"{user_input}"

---

Your task:
- Identify which meal (Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, or Dinner) the user wants to change.
- From the given meals list below, select a new meal of the same type as the one to replace.
- The new meal should have similar calories (within ±30 kcal) and follow the user's original dietary needs.
- Do not select the same meal that was originally suggested.
- Keep all other meals unchanged.

📄 Available Meals:
{meals_df.to_string(index=False)}

---

✅ Output Format (same as before):

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only change the requested meal. Keep everything else exactly as it is.
"""
    # response = model.generate_content(prompt)
    # return response.text.strip()
    response = llm.invoke(prompt)
    return response.text

template = """Answer the following questions as best you can. You have access to the following tools:

{tools}

Use the following format:

Question: the input question you must answer
Thought: you should always think about what to do
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought: {agent_scratchpad}"""

# Create the agent
agent = create_react_agent(
    llm=llm,
    tools=tools,
    prompt=PromptTemplate.from_template(template)
)

# Create the agent executor
agent_executor = AgentExecutor.from_agent_and_tools(
    agent=agent,
    tools=tools,
    verbose=True,
    memory=ConversationBufferMemory(memory_key="chat_history")
)
# ---------- HELPER: parse calories ----------
def parse_meal_calories(meal_text: str) -> dict:
    """Parse meal plan text to extract calories for each meal type"""
    calories = {}
    meal_types = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    
    for line in meal_text.splitlines():
        line = line.strip()
        if ':' in line and '(' in line:
            for meal_type in meal_types:
                if line.startswith(meal_type + ':'):
                    try:
                        # Extract calories from format like "(250 kcal" or "(250,"
                        match = re.search(r'\((\d+)', line)
                        if match:
                            calories[meal_type] = int(match.group(1))
                            break
                    except:
                        continue
    return calories

# Now other Streamlit commands can follow
st.title("🥗 Personalized Meal Plan Generator")
st.markdown("*Create customized meal plans based on your health goals and dietary preferences*")
st.divider()

# Create layout columns
col1, col2, col3 = st.columns([4, 2, 2], gap="large")

with col1:
    st.header("👤 Tell us about yourself")
    
    # User Input Form
    with st.form("user_info_form", clear_on_submit=False):
        # Personal Information Section
        st.subheader("📋 Personal Details")
        
        # Two column layout for personal info
        pers_col1, pers_col2 = st.columns(2)
        
        with pers_col1:
            name = st.text_input("Name", placeholder="Enter your name")
            age = st.number_input("Age", min_value=5, max_value=120, step=1, value=25)
            height = st.number_input("Height (cm)", min_value=100.0, max_value=250.0, step=1.0, value=170.0)
        
        with pers_col2:
            food_type = st.selectbox(
                "Preferred Cuisine",
                options=meals_df['Cuisine'].unique() if 'meals_df' in locals() else ['Indian', 'Continental', 'Asian', 'Mediterranean']
            )
            gender = st.selectbox("Gender", ["Male", "Female", "Other"])
            weight = st.number_input("Weight (kg)", min_value=20.0, max_value=300.0, step=1.0, value=70.0)

        st.divider()
        
        # Health Goals Section
        st.subheader("🎯 Health & Fitness Goals")
        health_goals = st.multiselect(
            "What are your primary health goals? (Select all that apply)",
            options=[
                "Weight management", "Energy boost", "Hormonal balance", "Bone & joint health",
                "Muscle gain", "Skin, hair & nails", "Gut health", "Immune support",
                "Fertility & reproductive health", "Menstrual wellness / Menopause support", "Other"
            ],
            help="Choose your main health objectives to get personalized recommendations"
        )

        st.divider()
        
        # Dietary Preferences Section  
        st.subheader("🍽️ Dietary Information")
        
        diet_col1, diet_col2 = st.columns(2)
        
        with diet_col1:
            diet = st.selectbox(
                "Dietary Pattern",
                options=["Omnivore", "Vegetarian", "Vegan", "Pescatarian", "Low-carb/Keto", "Gluten-free", "Jain/Sattvic", "Other"]
            )
            
            activity_level = st.selectbox(
                "Activity Level",
                options=[
                    "Little movement", 
                    "Walking, stretching everyday", 
                    "Exercise 2–3 times a week", 
                    "Daily training"
                ]
            )
        
        with diet_col2:
            food_allergies = st.multiselect(
                "Food Allergies & Intolerances",
                options=[
                    "Dairy (lactose or casein)", "Gluten (celiac or intolerance)", "Soy", "Peanuts", 
                    "Tree nuts", "Eggs", "Shellfish", "Fish", "Corn",
                    "Nightshades (tomatoes, peppers)", "Artificial sweeteners", "Caffeine", "Alcohol"
                ]
            )

        st.divider()
        
        # Health Conditions Section
        st.subheader("🏥 Health Conditions")
        disease = st.multiselect(
            "Do you currently manage any of these conditions?",
            options=[
                "Diabetes / Insulin resistance", "PCOS/PCOD", "High blood pressure / cholesterol", 
                "Thyroid imbalance (Hypo/Hyper)", "Hormonal Imbalance", "Digestive disorders (IBS, IBD, acidity)", 
                "Menstrual issues or menopause symptoms", "Autoimmune disease"
            ]
        )

        st.divider()
        
        # Supplement Preferences Section
        st.subheader("💊 Supplement Preferences (Optional)")
        supplement_preferences = st.multiselect(
            "What type of supplements do you prefer?",
            options=[
                "Herbal options", "Clinical-grade with scientific backing", "Vegan options",
                "Gluten-free options", "Gelatin-free options", "Soy-free options", "Corn-free options",
                "Cruelty-free options"
            ]
        )

        # Submit button
        st.markdown("")
        submitted = st.form_submit_button(
            "🚀 Generate My Meal Plan", 
            type="primary", 
            use_container_width=True
        )

    # Generate Meal Plan
    if submitted:
        user_profile = {
            "name": name,
            "age": age,
            "gender": gender,
            "height": height,
            "weight": weight,
            "health_goals": health_goals,
            "food_allergies": food_allergies,
            "diet": diet,
            "disease": disease,
            "activity_level": activity_level,
            "supplement_preferences": supplement_preferences
        }
        
        with st.spinner("🔄 Creating your personalized meal plan..."):
            # Mock meal plan generation (replace with your actual function)
            if 'meal_recommendation' in locals():
                meal_plan_text = meal_recommendation(user_profile, meals_df)
            else:
                meal_plan_text = f"""
## 🌟 Personalized Meal Plan for {name}

### Breakfast (400 kcal)
- Oatmeal with fresh berries and almonds
- Green tea

### Mid-Morning Snack (150 kcal)  
- Apple slices with almond butter

### Lunch (500 kcal)
- Quinoa salad with mixed vegetables
- Grilled chicken breast
- Lemon vinaigrette

### Afternoon Snack (100 kcal)
- Greek yogurt with honey

### Dinner (450 kcal)
- Baked salmon with herbs
- Roasted sweet potato and broccoli
- Brown rice

**Total Daily Calories: 1,600 kcal**
                """
                
        st.session_state['meal_plan'] = meal_plan_text
        
        # Initialize meal tracking
        if 'meal_completion' not in st.session_state:
            st.session_state['meal_completion'] = {}
        
        st.success("✅ Your meal plan is ready!")

    # Display Meal Plan
    if 'meal_plan' in st.session_state:
        st.divider()
        st.header("🍽️ Your Personalized Meal Plan")
        st.markdown(st.session_state['meal_plan'])

    # Meal Modification Section
    if 'meal_plan' in st.session_state:
        st.divider()
        st.header("🔄 Modify Your Plan")
        
        with st.form("modification_form"):
            st.write("Need changes to your meal plan? Let me know!")
            user_input = st.text_area(
                "Describe the changes you'd like:",
                placeholder="e.g., 'Replace chicken with tofu', 'Add more vegetables', 'Reduce calories'"
            )
            change_submitted = st.form_submit_button("✨ Apply Changes", type="primary")
        
        if change_submitted and user_input:
            with st.spinner("🔄 Updating your meal plan..."):
                # Mock modification (replace with your actual function)
                if 'meal_change_according_to_user' in locals():
                    updated_meal = meal_change_according_to_user(user_input, st.session_state['meal_plan'], meals_df)
                else:
                    updated_meal = st.session_state['meal_plan'] + f"\n\n---\n**🔄 Updated based on your request:** {user_input}"
            
            st.session_state['meal_plan'] = updated_meal
            st.success("✅ Your meal plan has been updated!")
            st.markdown(updated_meal)

# Progress Tracking Column
with col2:
    st.header("📊 Daily Progress")
    
    if 'meal_plan' in st.session_state:
        # Mock calorie parsing function (replace with your actual function)
        def parse_meal_calories(meal_plan):
            return {
                "Breakfast": 400,
                "Mid-Morning Snack": 150,
                "Lunch": 500,
                "Afternoon Snack": 100,
                "Dinner": 450
            }
        
        calories_map = parse_meal_calories(st.session_state['meal_plan'])
        
        if calories_map:
            total_required = sum(calories_map.values())
            
            # Initialize tracking
            if 'meal_completion' not in st.session_state:
                st.session_state['meal_completion'] = {meal: False for meal in calories_map.keys()}
            
            st.subheader("✅ Track Your Meals")
            st.write("Check off meals as you complete them:")
            
            consumed_calories = 0
            
            # Meal checkboxes
            for meal_type, cal in calories_map.items():
                completed = st.checkbox(
                    f"{meal_type} ({cal} kcal)", 
                    key=f"meal_{meal_type}",
                    value=st.session_state['meal_completion'].get(meal_type, False)
                )
                st.session_state['meal_completion'][meal_type] = completed
                
                if completed:
                    consumed_calories += cal
            
            # Progress metrics
            remaining_calories = max(total_required - consumed_calories, 0)
            progress_percentage = (consumed_calories / total_required) * 100 if total_required > 0 else 0
            
            st.divider()
            
            # Display metrics using Streamlit's metric component
            metric_col1, metric_col2 = st.columns(2)
            with metric_col1:
                st.metric("Consumed", f"{consumed_calories} kcal")
            with metric_col2:
                st.metric("Remaining", f"{remaining_calories} kcal")
            
            st.metric("Progress", f"{progress_percentage:.1f}%")
            
            # Progress bar
            st.progress(progress_percentage / 100)
            
            # Status messages
            if consumed_calories >= total_required:
                st.success("🎉 Daily goal completed!")
            elif consumed_calories > 0:
                st.info(f"📈 {progress_percentage:.1f}% complete")
            else:
                st.warning("🍽️ Start tracking your meals!")
            
            # Visual chart
            if consumed_calories > 0 or remaining_calories > 0:
                st.subheader("📈 Visual Progress")
                
                fig, ax = plt.subplots(figsize=(8, 6))
                
                labels = []
                sizes = []
                colors = []
                
                if consumed_calories > 0:
                    labels.append(f'Consumed\n{consumed_calories} kcal')
                    sizes.append(consumed_calories)
                    colors.append('#00C851')
                
                if remaining_calories > 0:
                    labels.append(f'Remaining\n{remaining_calories} kcal')
                    sizes.append(remaining_calories)
                    colors.append('#ffbb33')
                
                wedges, texts, autotexts = ax.pie(
                    sizes, 
                    labels=labels, 
                    autopct='%1.1f%%', 
                    startangle=90,
                    colors=colors
                )
                
                ax.set_title('Daily Calorie Progress', fontsize=14, fontweight='bold')
                ax.axis('equal')
                
                st.pyplot(fig)
                plt.close()
            
        else:
            st.error("Unable to parse meal plan calories")
    else:
        st.info("🍽️ Generate a meal plan to start tracking your progress!")

# Adjustment Helper Column
with col3:
    st.header("🔧 Smart Adjustments")
    
    if 'meal_plan' in st.session_state:
        st.subheader("Overate Something?")
        st.write("Don't worry! Let's adjust your plan.")
        
        with st.form("adjustment_form"):
            over_dish = st.text_input(
                "What did you overeat?", 
                placeholder="e.g., rice, pasta, dessert"
            )
            
            over_qty = st.number_input(
                "How many extra servings?", 
                min_value=1, 
                max_value=10,
                step=1,
                value=1
            )
            
            adjust_submitted = st.form_submit_button("🔄 Get Suggestions", type="secondary")
        
        if adjust_submitted and over_dish:
            with st.spinner("🔄 Calculating adjustments..."):
                # Mock adjustment function (replace with your actual function)
                if 'adjust_meal_after_overeating' in locals():
                    result = adjust_meal_after_overeating(over_dish, over_qty, meals_df, st.session_state['meal_plan'])
                else:
                    result = f"""
### 💡 Smart Adjustments

You overate **{over_dish}** ({over_qty} extra servings).

**Immediate Actions:**
- Drink extra water
- Take a 10-minute walk

**Today's Adjustments:**
- Reduce next meal by 15%
- Skip or halve your next snack
- Add 5 minutes of activity

**Tomorrow's Plan:**
- Start with a lighter breakfast
- Focus on protein and vegetables
- Stay hydrated throughout the day
                    """
            
            st.success("✅ Adjustment plan ready!")
            st.markdown(result)
        
        st.divider()
        
        # Quick tips
        st.subheader("💡 Quick Tips")
        
        tips = [
            "🥤 Drink water before meals",
            "🥗 Fill half your plate with vegetables", 
            "🚶‍♀️ Take a 5-minute walk after eating",
            "⏰ Eat mindfully and slowly",
            "😴 Get adequate sleep for better metabolism"
        ]
        
        for tip in tips:
            st.write(tip)
            
    else:
        st.info("🍽️ Generate a meal plan first to access adjustment tools!")
        
        # General wellness tips while waiting
        st.subheader("🌟 Wellness Tips")
        st.write("While you create your meal plan:")
        
        general_tips = [
            "💧 Aim for 8 glasses of water daily",
            "🥬 Include colorful vegetables",
            "🥜 Add healthy fats like nuts and seeds",
            "🏃‍♀️ Move your body daily",
            "😌 Practice mindful eating"
        ]
        
        for tip in general_tips:
            st.write(tip)

# Footer
st.divider()
st.markdown("---")

# Summary footer
footer_col1, footer_col2, footer_col3 = st.columns(3)

with footer_col1:
    st.markdown("### 🎯 Your Goals")
    st.write("Personalized nutrition for better health")

with footer_col2:
    st.markdown("### 📊 Track Progress") 
    st.write("Monitor your daily meal completion")

with footer_col3:
    st.markdown("### 🔄 Stay Flexible")
    st.write("Adjust your plan as needed")

st.markdown("---")
st.markdown(
    "<div style='text-align: center;'><b>🌟 Your journey to better health starts here! 🌟</b></div>", 
    unsafe_allow_html=True
)
```

```python
import streamlit as st
import pandas as pd
import os
from langchain.agents import Tool, AgentExecutor, create_react_agent
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
import google.generativeai as genai
import matplotlib.pyplot as plt
import re
from dotenv import load_dotenv

st.set_page_config(page_title="🥗 Personalized Meal Plan Generator", layout="wide")

load_dotenv()

# ---------- CONFIGURE API ----------
# Initialize the Search Tool
search = DuckDuckGoSearchRun()
tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Useful for searching information on the internet. Use this when you need to find current or factual information about food calories."
    )
]

# Initialize the language model
llm = ChatGroq(
    temperature=0.7,
    model="llama3-8b-8192",
    api_key=os.getenv("GROQ_API_KEY"),
)

api_key = "AIzaSyBMdmH6lJdWLtqQ4YBB8bkqXaGukQm_47s"
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.0-flash-001')

# ---------- LOAD DATASETS ----------
@st.cache_data
def load_meals(csv_path: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])
    return df

@st.cache_data
def load_nutrition_data(csv_path: str) -> pd.DataFrame:
    """Load the nutrition data CSV for calorie lookup"""
    try:
        df = pd.read_csv(csv_path)
        return df
    except FileNotFoundError:
        st.warning("nutrition_data.csv not found. Agent will be used for all unknown foods.")
        return pd.DataFrame()

meals_df = load_meals("meal.csv")
nutrition_df = load_nutrition_data("nutrition_data.csv")

# ---------- CALORIE LOOKUP FUNCTIONS ----------
def search_calories_in_nutrition_data(food_item: str) -> dict:
    """
    Search for food item in nutrition_data.csv
    Returns: dict with 'found', 'calories', 'portion' keys
    """
    if nutrition_df.empty:
        return {'found': False, 'calories': None, 'portion': None}
    
    # Normalize the food item name for better matching
    food_item_lower = food_item.lower().strip()
    
    # Search in the Food Item column (case-insensitive)
    matches = nutrition_df[nutrition_df['Food Item'].str.lower().str.contains(food_item_lower, na=False)]
    
    if not matches.empty:
        # Take the first match
        match = matches.iloc[0]
        return {
            'found': True,
            'calories': match['Calories'],
            'portion': match['Portion'],
            'food_name': match['Food Item']
        }
    
    return {'found': False, 'calories': None, 'portion': None}

def get_calories_from_agent(food_item: str) -> dict:
    """
    Use agent to search for calorie information online
    Returns: dict with 'found', 'calories', 'portion' keys
    """
    # Create the agent
    template = """Answer the following questions as best you can. You have access to the following tools:

{tools}

Use the following format:

Question: the input question you must answer
Thought: you should always think about what to do
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought: {agent_scratchpad}"""

    agent = create_react_agent(
        llm=llm,
        tools=tools,
        prompt=PromptTemplate.from_template(template)
    )

    agent_executor = AgentExecutor.from_agent_and_tools(
        agent=agent,
        tools=tools,
        verbose=False,
        memory=ConversationBufferMemory(memory_key="chat_history"),
        max_iterations=3
    )
    
    try:
        query = f"What are the calories in {food_item}? Provide just the calorie number per standard serving portion."
        response = agent_executor.invoke({"input": query})
        
        # Extract calories from response
        response_text = response.get('output', '')
        
        # Try to extract calorie number from response
        calorie_match = re.search(r'(\d+)\s*(?:cal|kcal|calories)', response_text.lower())
        
        if calorie_match:
            calories = int(calorie_match.group(1))
            return {
                'found': True,
                'calories': calories,
                'portion': 'standard serving',
                'source': 'agent_search'
            }
        
    except Exception as e:
        st.error(f"Agent search failed: {str(e)}")
    
    return {'found': False, 'calories': None, 'portion': None}

def get_food_calories(food_item: str) -> dict:
    """
    Main function to get calories for any food item
    First checks nutrition_data.csv, then uses agent as fallback
    """
    # Step 1: Check nutrition data CSV
    nutrition_result = search_calories_in_nutrition_data(food_item)
    
    if nutrition_result['found']:
        st.info(f"✅ Found '{food_item}' in nutrition database")
        return nutrition_result
    
    # Step 2: Use agent as fallback
    st.info(f"🔍 '{food_item}' not found in database. Searching online...")
    agent_result = get_calories_from_agent(food_item)
    
    if agent_result['found']:
        st.success(f"✅ Found calorie information for '{food_item}' online")
        return agent_result
    else:
        st.error(f"❌ Could not find calorie information for '{food_item}'")
        return {'found': False, 'calories': None, 'portion': None}

# ---------- BMI and BMR functions ----------
def classify_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)

def calculate_bmr(weight_kg, height_cm, age, gender):
    if gender.lower().startswith('m'):
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age + 5, 1)
    else:
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age - 161, 1)

# ---------- MEAL RECOMMENDATION ----------
def meal_recommendation(user_profile: dict, meals_df: pd.DataFrame) -> str:
    bmi_value = classify_bmi(user_profile['weight'], user_profile['height'])
    bmr = calculate_bmr(
        user_profile['weight'], user_profile['height'], user_profile['age'], user_profile['gender']
    )

    prompt = f"""
You are an expert dietitian assistant.

User Profile:
- Name: {user_profile['name']}
- Age: {user_profile['age']}
- Gender: {user_profile['gender']}
- Height: {user_profile['height']} cm
- Weight: {user_profile['weight']} kg
- BMI: {bmi_value}
- BMR: {bmr} kcal/day
- Health Goals: {', '.join(user_profile['health_goals'])}
- Food Allergies: {', '.join(user_profile['food_allergies'])}
- Dietary Pattern: {user_profile['diet']}
- Disease: {', '.join(user_profile['disease'])}
- Activity Level: {user_profile['activity_level']}
- Supplement Preferences: {', '.join(user_profile['supplement_preferences'])}

---

Task:
- Create a personalized meal plan with exactly 5 meals per day: Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, and Dinner.
- Choose **only** from this list of meals given below. Do not invent new meals.
- Each meal in the list has type (Breakfast, Snack, Lunch, Dinner), calories, protein, carbs, and fats.
- Consider user's BMI:
  * If BMI <18.5: Recommend meals above BMR (weight gain).
  * If BMI >=25: Recommend meals below BMR (weight loss).
  * If BMI 18.5–25: Recommend around BMR (maintenance).

---

📄 **Available Meals**:

{meals_df.to_string(index=False)}

---

✅ Format your answer like this:

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only pick from the provided list. Keep it personalized and explain briefly why you chose each.
"""

    # response = llm.invoke(prompt)
    # return response.content
    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- MEAL MODIFICATION FUNCTION ----------
def handle_food_intake_modification(user_input: str, current_meal_plan: str) -> str:
    """
    Handle when user inputs food not in the suggested meal plan
    """
    # Extract food items mentioned by user
    food_items = extract_food_items_from_input(user_input)
    
    if not food_items:
        return "I couldn't identify specific food items in your request. Please mention the specific food you want to add or replace."
    
    # Get calorie information for each food item
    food_calorie_info = {}
    for food in food_items:
        calorie_info = get_food_calories(food)
        food_calorie_info[food] = calorie_info
    
    # Create modified meal plan prompt
    calorie_info_text = ""
    for food, info in food_calorie_info.items():
        if info['found']:
            calorie_info_text += f"- {food}: {info['calories']} calories per {info['portion']}\n"
    
    if not calorie_info_text:
        return "I couldn't find calorie information for the foods you mentioned. Please try with different food items."
    
    modification_prompt = f"""
You are an expert dietitian assistant.

Current meal plan:
{current_meal_plan}

User wants to modify the plan with: {user_input}

Available calorie information for user's food items:
{calorie_info_text}

Task:
- Modify the meal plan according to user's request
- Replace or add the mentioned foods appropriately
- Maintain similar total daily calories
- Keep the same format as the original plan

Provide the updated meal plan in the same format:
Breakfast: [Meal Name] (calories, protein, carbs, fats) - Justification.
Mid-Morning Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.
Lunch: [Meal Name] (calories, protein, carbs, fats) - Justification.
Afternoon Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.
Dinner: [Meal Name] (calories, protein, carbs, fats) - Justification.
"""
    
    # response = llm.invoke(modification_prompt)
    # return response.content
    response = model.generate_content(modification_prompt)
    return response.text.strip()

def extract_food_items_from_input(user_input: str) -> list:
    """
    Extract food items from user input
    This is a simple implementation - you can make it more sophisticated
    """
    # Common food-related keywords that might indicate food items
    food_keywords = []
    
    # Split input into words and look for potential food items
    words = user_input.lower().split()
    
    # Simple approach: look for common food patterns
    potential_foods = []
    
    # Look for phrases like "add pizza", "replace with burger", etc.
    for i, word in enumerate(words):
        if word in ['add', 'replace', 'with', 'want', 'like', 'prefer']:
            # Check next few words for food items
            for j in range(i+1, min(i+4, len(words))):
                if j < len(words):
                    potential_foods.append(words[j])
    
    # Also check against nutrition database for any matches
    if not nutrition_df.empty:
        for word in words:
            matches = nutrition_df[nutrition_df['Food Item'].str.lower().str.contains(word, na=False)]
            if not matches.empty:
                potential_foods.append(word)
    
    # Remove duplicates and common non-food words
    non_food_words = {'with', 'and', 'the', 'a', 'an', 'some', 'more', 'less', 'want', 'like', 'prefer', 'add', 'replace'}
    food_items = list(set([food for food in potential_foods if food not in non_food_words and len(food) > 2]))
    
    return food_items

# ---------- MEAL CONFIGURATION BY USER (UPDATED) ----------
def meal_change_according_to_user(user_input: str, recommended_meal: str, meals_df: pd.DataFrame) -> str:
    prompt = f"""
You are an expert dietitian assistant.

Below is the meal plan previously recommended to the user:

{recommended_meal}

The user has requested the following change:
"{user_input}"

---

Your task:
- Identify which meal (Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, or Dinner) the user wants to change.
- From the given meals list below, select a new meal of the same type as the one to replace.
- The new meal should have similar calories (within ±30 kcal) and follow the user's original dietary needs.
- Do not select the same meal that was originally suggested.
- Keep all other meals unchanged.

📄 Available Meals:
{meals_df.to_string(index=False)}

---

✅ Output Format (same as before):

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only change the requested meal. Keep everything else exactly as it is.
"""

    # response = llm.invoke(prompt)
    # return response.content
    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- HELPER: parse calories ----------
def parse_meal_calories(meal_text: str) -> dict:
    """Parse meal plan text to extract calories for each meal type"""
    calories = {}
    meal_types = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    for line in meal_text.splitlines():
        line = line.strip()
        if ':' in line and '(' in line:
            for meal_type in meal_types:
                if line.startswith(meal_type + ':'):
                    try:
                        # Extract calories from format like "(250 kcal" or "(250,"
                        match = re.search(r'\((\d+)', line)
                        if match:
                            calories[meal_type] = int(match.group(1))
                            break
                    except:
                        continue
    return calories

# ---------- ENHANCED CALORIE CALCULATION WITH QUANTITY ----------
def parse_food_with_quantity(user_input: str) -> dict:
    """
    Parse user input to extract food item and quantity
    Examples: "2 slices of pizza", "1 cup rice", "200g chicken"
    Returns: {'food': str, 'quantity': float, 'unit': str, 'raw_input': str}
    """
    user_input = user_input.strip().lower()
    
    # Common quantity patterns
    quantity_patterns = [
        r'(\d+(?:\.\d+)?)\s*(cups?|cup)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(slices?|slice)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(pieces?|piece)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*([kg|g|grams?|kilograms?])\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(bowls?|bowl)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(plates?|plate)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(.+)',  # Just number and food
    ]
    
    for pattern in quantity_patterns:
        match = re.search(pattern, user_input)
        if match:
            if len(match.groups()) == 3:
                quantity, unit, food = match.groups()
                return {
                    'food': food.strip(),
                    'quantity': float(quantity),
                    'unit': unit.strip(),
                    'raw_input': user_input
                }
            elif len(match.groups()) == 2:
                quantity, food = match.groups()
                return {
                    'food': food.strip(),
                    'quantity': float(quantity),
                    'unit': 'serving',
                    'raw_input': user_input
                }
    
    # If no quantity found, assume 1 serving
    return {
        'food': user_input,
        'quantity': 1.0,
        'unit': 'serving',
        'raw_input': user_input
    }

def calculate_adjusted_calories(food_info: dict, base_calories: int) -> int:
    """
    Calculate adjusted calories based on quantity and unit
    """
    quantity = food_info['quantity']
    unit = food_info['unit']
    
    # Define multipliers for different units
    unit_multipliers = {
        'serving': 1.0,
        'cup': 1.0,
        'slice': 0.8,  # Assuming slice is smaller than full serving
        'piece': 0.8,
        'bowl': 1.2,   # Bowl is usually larger
        'plate': 1.5,  # Plate is usually larger
        'g': 0.01,     # Per gram (assuming base is per 100g)
        'gram': 0.01,
        'grams': 0.01,
        'kg': 10.0,    # Per kg
        'kilogram': 10.0,
        'kilograms': 10.0,
    }
    
    multiplier = unit_multipliers.get(unit, 1.0)
    adjusted_calories = int(base_calories * quantity * multiplier)
    
    return adjusted_calories

def get_food_calories_with_quantity(user_input: str) -> dict:
    """
    Enhanced function to get calories considering quantity
    """
    # Parse food and quantity
    food_info = parse_food_with_quantity(user_input)
    
    # Get base calories for the food item
    calorie_result = get_food_calories(food_info['food'])
    
    if calorie_result['found']:
        # Calculate adjusted calories based on quantity
        adjusted_calories = calculate_adjusted_calories(food_info, calorie_result['calories'])
        
        return {
            'found': True,
            'food': food_info['food'],
            'quantity': food_info['quantity'],
            'unit': food_info['unit'],
            'base_calories': calorie_result['calories'],
            'adjusted_calories': adjusted_calories,
            'portion': calorie_result['portion'],
            'raw_input': food_info['raw_input']
        }
    
    return {'found': False, 'raw_input': user_input}

# ---------- MEAL ADJUSTMENT BASED ON CALORIE DIFFERENCE ----------
def adjust_next_meals(consumed_food_info: dict, current_meal_plan: str, meal_eaten: str) -> str:
    """
    Adjust the next meals based on calorie difference from what was consumed vs planned
    """
    # Parse current meal plan to get planned calories
    planned_calories = parse_meal_calories(current_meal_plan)
    
    if meal_eaten not in planned_calories:
        return "Could not identify which meal to adjust. Please specify (Breakfast, Lunch, Dinner, etc.)"
    
    # Calculate calorie difference
    planned_meal_calories = planned_calories[meal_eaten]
    consumed_calories = consumed_food_info['adjusted_calories']
    calorie_difference = consumed_calories - planned_meal_calories
    
    # Get meal order to determine next meals
    meal_order = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    
    try:
        current_meal_index = meal_order.index(meal_eaten)
        remaining_meals = meal_order[current_meal_index + 1:]
    except ValueError:
        remaining_meals = []
    
    if not remaining_meals:
        return f"""
✅ **Meal Update Complete**

You consumed: **{consumed_food_info['food'].title()}** ({consumed_food_info['quantity']} {consumed_food_info['unit']})
- Calories: {consumed_calories} kcal
- Planned: {planned_meal_calories} kcal
- Difference: {'+' if calorie_difference > 0 else ''}{calorie_difference} kcal

Since this was your last meal of the day, no further adjustments needed. 
{('Consider lighter meals tomorrow if you exceeded your target.' if calorie_difference > 0 else 'Great job staying within your calorie goals!')}
        """
    
    # Create adjustment prompt
    adjustment_prompt = f"""
You are an expert dietitian assistant.

Current meal plan:
{current_meal_plan}

**Meal Adjustment Needed:**
- User consumed: {consumed_food_info['food']} ({consumed_food_info['quantity']} {consumed_food_info['unit']}) = {consumed_calories} kcal
- Originally planned for {meal_eaten}: {planned_meal_calories} kcal  
- Calorie difference: {calorie_difference} kcal {'(excess)' if calorie_difference > 0 else '(deficit)'}

**Remaining meals to adjust:** {', '.join(remaining_meals)}

**Task:**
1. Keep the consumed meal as: "{meal_eaten}: {consumed_food_info['food'].title()} ({consumed_food_info['quantity']} {consumed_food_info['unit']}) ({consumed_calories} kcal) - User consumed this instead"
2. Adjust the remaining meals ({', '.join(remaining_meals)}) to compensate for the {calorie_difference} kcal difference
3. If excess calories: Choose lower-calorie alternatives from the meal database
4. If deficit calories: Choose slightly higher-calorie alternatives or add healthy additions
5. Distribute the adjustment across remaining meals
6. Select only from the provided meals database

Available Meals Database:
{meals_df.to_string(index=False)}

**Output the complete adjusted meal plan in this format:**

Breakfast: [Meal Name] (calories, protein, carbs, fats) - Justification.
Mid-Morning Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.  
Lunch: [Meal Name] (calories, protein, carbs, fats) - Justification.
Afternoon Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.
Dinner: [Meal Name] (calories, protein, carbs, fats) - Justification.
"""
    
    response = model.generate_content(adjustment_prompt)
    return response.text.strip()

# ---------- UPDATED UI SECTION FOR ADJUSTMENT ----------
# Replace the existing adjustment form in col3 with this enhanced version:

def render_enhanced_adjustment_section():
    """Enhanced adjustment section UI"""
    st.subheader("🍽️ Ate Something Different?")
    st.write("Tell me what you ate and I'll adjust your remaining meals!")
    
    with st.form("enhanced_adjustment_form"):
        # Meal selection
        meal_eaten = st.selectbox(
            "Which meal did you just have?",
            options=["Breakfast", "Mid-Morning Snack", "Lunch", "Afternoon Snack", "Dinner"],
            help="Select the meal you want to replace"
        )
        
        # Food input with quantity
        food_consumed = st.text_input(
            "What did you eat? (Include quantity)",
            placeholder="e.g., 2 slices pizza, 1 cup rice, 200g chicken",
            help="Be specific about quantity for accurate calorie calculation"
        )
        
        # Examples
        with st.expander("💡 Quantity Examples"):
            st.write("""
            **Valid formats:**
            - 2 slices of pizza
            - 1 cup rice  
            - 200g chicken breast
            - 1 bowl pasta
            - 3 pieces samosa
            - 1 plate biryani
            """)
        
        adjust_submitted = st.form_submit_button("🔄 Adjust My Meal Plan", type="primary")
        
        if adjust_submitted and food_consumed:
            with st.spinner("🔄 Calculating calories and adjusting your plan..."):
                # Get food info with quantity
                food_info = get_food_calories_with_quantity(food_consumed)
                
                if food_info['found']:
                    # Show what was consumed
                    st.success(f"✅ **Food Logged:** {food_info['food'].title()}")
                    
                    col1, col2 = st.columns(2)
                    with col1:
                        st.metric("Quantity", f"{food_info['quantity']} {food_info['unit']}")
                    with col2:
                        st.metric("Total Calories", f"{food_info['adjusted_calories']} kcal")
                    
                    # Adjust the meal plan
                    adjusted_plan = adjust_next_meals(food_info, st.session_state['meal_plan'], meal_eaten)
                    st.session_state['meal_plan'] = adjusted_plan
                    
                    st.success("✅ Your meal plan has been adjusted!")
                    st.rerun()
                else:
                    st.error("Could not find calorie information for this food item. Please try with a different food name.")

# ---------- STREAMLIT UI (Rest remains same) ----------
st.title("🥗 Personalized Meal Plan Generator")
st.markdown("*Create customized meal plans based on your health goals and dietary preferences*")
st.divider()

# Create layout columns
col1, col2, col3 = st.columns([4, 2, 2], gap="large")

with col1:
    st.header("👤 Tell us about yourself")
    
    # User Input Form
    with st.form("user_info_form", clear_on_submit=False):
        # Personal Information Section
        st.subheader("📋 Personal Details")
        
        # Two column layout for personal info
        pers_col1, pers_col2 = st.columns(2)
        
        with pers_col1:
            name = st.text_input("Name", placeholder="Enter your name")
            age = st.number_input("Age", min_value=5, max_value=120, step=1, value=25)
            height = st.number_input("Height (cm)", min_value=100.0, max_value=250.0, step=1.0, value=170.0)
        
        with pers_col2:
            food_type = st.selectbox(
                "Preferred Cuisine",
                options=meals_df['Cuisine'].unique() if 'Cuisine' in meals_df.columns else ['Indian', 'Continental', 'Asian', 'Mediterranean']
            )
            gender = st.selectbox("Gender", ["Male", "Female", "Other"])
            weight = st.number_input("Weight (kg)", min_value=20.0, max_value=300.0, step=1.0, value=70.0)
        
        st.divider()
        
        # Health Goals Section
        st.subheader("🎯 Health & Fitness Goals")
        health_goals = st.multiselect(
            "What are your primary health goals? (Select all that apply)",
            options=[
                "Weight management", "Energy boost", "Hormonal balance", "Bone & joint health",
                "Muscle gain", "Skin, hair & nails", "Gut health", "Immune support",
                "Fertility & reproductive health", "Menstrual wellness / Menopause support", "Other"
            ],
            help="Choose your main health objectives to get personalized recommendations"
        )
        
        st.divider()
        
        # Dietary Preferences Section
        st.subheader("🍽️ Dietary Information")
        diet_col1, diet_col2 = st.columns(2)
        
        with diet_col1:
            diet = st.selectbox(
                "Dietary Pattern",
                options=["Omnivore", "Vegetarian", "Vegan", "Pescatarian", "Low-carb/Keto", "Gluten-free", "Jain/Sattvic", "Other"]
            )
            
            activity_level = st.selectbox(
                "Activity Level",
                options=[
                    "Little movement",
                    "Walking, stretching everyday",
                    "Exercise 2–3 times a week",
                    "Daily training"
                ]
            )
        
        with diet_col2:
            food_allergies = st.multiselect(
                "Food Allergies & Intolerances",
                options=[
                    "Dairy (lactose or casein)", "Gluten (celiac or intolerance)", "Soy", "Peanuts",
                    "Tree nuts", "Eggs", "Shellfish", "Fish", "Corn",
                    "Nightshades (tomatoes, peppers)", "Artificial sweeteners", "Caffeine", "Alcohol"
                ]
            )
        
        st.divider()
        
        # Health Conditions Section
        st.subheader("🏥 Health Conditions")
        disease = st.multiselect(
            "Do you currently manage any of these conditions?",
            options=[
                "Diabetes / Insulin resistance", "PCOS/PCOD", "High blood pressure / cholesterol",
                "Thyroid imbalance (Hypo/Hyper)", "Hormonal Imbalance", "Digestive disorders (IBS, IBD, acidity)",
                "Menstrual issues or menopause symptoms", "Autoimmune disease"
            ]
        )
        
        st.divider()
        
        # Supplement Preferences Section
        st.subheader("💊 Supplement Preferences (Optional)")
        supplement_preferences = st.multiselect(
            "What type of supplements do you prefer?",
            options=[
                "Herbal options", "Clinical-grade with scientific backing", "Vegan options",
                "Gluten-free options", "Gelatin-free options", "Soy-free options", "Corn-free options",
                "Cruelty-free options"
            ]
        )
        
        # Submit button
        st.markdown("")
        submitted = st.form_submit_button(
            "🚀 Generate My Meal Plan",
            type="primary",
            use_container_width=True
        )

# Generate Meal Plan
if submitted:
    user_profile = {
        "name": name,
        "age": age,
        "gender": gender,
        "height": height,
        "weight": weight,
        "health_goals": health_goals,
        "food_allergies": food_allergies,
        "diet": diet,
        "disease": disease,
        "activity_level": activity_level,
        "supplement_preferences": supplement_preferences
    }

    with st.spinner("🔄 Creating your personalized meal plan..."):
        meal_plan_text = meal_recommendation(user_profile, meals_df)
        st.session_state['meal_plan'] = meal_plan_text

    # Initialize meal tracking
    if 'meal_completion' not in st.session_state:
        st.session_state['meal_completion'] = {}

    st.success("✅ Your meal plan is ready!")

# Display Meal Plan
if 'meal_plan' in st.session_state:
    st.divider()
    st.header("🍽️ Your Personalized Meal Plan")
    st.markdown(st.session_state['meal_plan'])

    # Meal Modification Section
    st.divider()
    st.header("🔄 Modify Your Plan")
    
    with st.form("modification_form"):
        st.write("Need changes to your meal plan? Let me know!")
        st.info("💡 You can ask for foods not in our meal database - I'll search for their calorie information online!")
        
        user_input = st.text_area(
            "Describe the changes you'd like:",
            placeholder="e.g., 'Replace chicken with tofu', 'Add pizza to lunch', 'I want burger instead'"
        )

        change_submitted = st.form_submit_button("✨ Apply Changes", type="primary")

        if change_submitted and user_input:
            with st.spinner("🔄 Updating your meal plan..."):
                updated_meal = meal_change_according_to_user(user_input, st.session_state['meal_plan'], meals_df)
                st.session_state['meal_plan'] = updated_meal
                st.success("✅ Your meal plan has been updated!")
                st.rerun()

# Progress Tracking Column (Rest of the code remains the same)
with col2:
    st.header("📊 Daily Progress")
    if 'meal_plan' in st.session_state:
        calories_map = parse_meal_calories(st.session_state['meal_plan'])
        
        if calories_map:
            total_required = sum(calories_map.values())
            
            # Initialize tracking
            if 'meal_completion' not in st.session_state:
                st.session_state['meal_completion'] = {meal: False for meal in calories_map.keys()}

            st.subheader("✅ Track Your Meals")
            st.write("Check off meals as you complete them:")

            consumed_calories = 0

            # Meal checkboxes
            for meal_type, cal in calories_map.items():
                completed = st.checkbox(
                    f"{meal_type} ({cal} kcal)",
                    key=f"meal_{meal_type}",
                    value=st.session_state['meal_completion'].get(meal_type, False)
                )

                st.session_state['meal_completion'][meal_type] = completed
                if completed:
                    consumed_calories += cal

            # Progress metrics
            remaining_calories = max(total_required - consumed_calories, 0)
            progress_percentage = (consumed_calories / total_required) * 100 if total_required > 0 else 0

            st.divider()

            # Display metrics using Streamlit's metric component
            metric_col1, metric_col2 = st.columns(2)

            with metric_col1:
                st.metric("Consumed", f"{consumed_calories} kcal")

            with metric_col2:
                st.metric("Remaining", f"{remaining_calories} kcal")

            st.metric("Progress", f"{progress_percentage:.1f}%")

            # Progress bar
            st.progress(progress_percentage / 100)

            # Status messages
            if consumed_calories >= total_required:
                st.success("🎉 Daily goal completed!")
            elif consumed_calories > 0:
                st.info(f"📈 {progress_percentage:.1f}% complete")
            else:
                st.warning("🍽️ Start tracking your meals!")

            # Visual chart
            if consumed_calories > 0 or remaining_calories > 0:
                st.subheader("📈 Visual Progress")
                fig, ax = plt.subplots(figsize=(8, 6))
                labels = []
                sizes = []
                colors = []

                if consumed_calories > 0:
                    labels.append(f'Consumed\n{consumed_calories} kcal')
                    sizes.append(consumed_calories)
                    colors.append('#00C851')

                if remaining_calories > 0:
                    labels.append(f'Remaining\n{remaining_calories} kcal')
                    sizes.append(remaining_calories)
                    colors.append('#ffbb33')

                wedges, texts, autotexts = ax.pie(
                    sizes,
                    labels=labels,
                    autopct='%1.1f%%',
                    startangle=90,
                    colors=colors
                )

                ax.set_title('Daily Calorie Progress', fontsize=14, fontweight='bold')
                ax.axis('equal')
                st.pyplot(fig)
                plt.close()

        else:
            st.error("Unable to parse meal plan calories")
    else:
        st.info("🍽️ Generate a meal plan to start tracking your progress!")

# Adjustment Helper Column (Rest remains the same)
with col3:
    st.header("🔧 Smart Adjustments")
    
    if 'meal_plan' in st.session_state:
        render_enhanced_adjustment_section()
        
        # st.subheader("Ate Something Different?")
        # st.write("No worries! I can help adjust your plan.")
        
        # with st.form("adjustment_form"):
        #     different_food = st.text_input(
        #         "What did you eat instead?",
        #         placeholder="e.g., pizza, burger, pasta"
        #     )

        #     adjust_submitted = st.form_submit_button("🔍 Get Calorie Info", type="secondary")

        #     if adjust_submitted and different_food:
        #         with st.spinner("🔄 Looking up calorie information..."):
        #             calorie_info = get_food_calories(different_food)
                    
        #             if calorie_info['found']:
        #                 st.success(f"**{different_food.title()}**")
        #                 st.write(f"📊 **Calories:** {calorie_info['calories']} per {calorie_info['portion']}")
                        
        #                 if 'source' in calorie_info and calorie_info['source'] == 'agent_search':
        #                     st.info("ℹ️ Information found online")
        #                 else:
        #                     st.info("ℹ️ Information from nutrition database")
                            
        #                 # Provide adjustment suggestions
        #                 st.write("**💡 Adjustment Suggestions:**")
        #                 st.write("- Compare with your planned meal calories")
        #                 st.write("- Adjust your next meal if needed")
        #                 st.write("- Stay hydrated!")
        #             else:
        #                 st.error("Could not find calorie information for this food item.")

        st.divider()

        # Quick tips
        st.subheader("💡 Quick Tips")
        tips = [
            "🥤 Drink water before meals",
            "🥗 Fill half your plate with vegetables",
            "🚶‍♀️ Take a 5-minute walk after eating",
            "⏰ Eat mindfully and slowly",
            "😴 Get adequate sleep for better metabolism"
        ]

        for tip in tips:
            st.write(tip)
    else:
        st.info("🍽️ Generate a meal plan first to access adjustment tools!")

        # General wellness tips while waiting
        st.subheader("🌟 Wellness Tips")
        st.write("While you create your meal plan:")
        
        general_tips = [
            "💧 Aim for 8 glasses of water daily",
            "🥬 Include colorful vegetables",
            "🥜 Add healthy fats like nuts and seeds",
            "🏃‍♀️ Move your body daily",
            "😌 Practice mindful eating"
        ]

        for tip in general_tips:
            st.write(tip)

# Footer
st.divider()
st.markdown("---")

# Summary footer
footer_col1, footer_col2, footer_col3 = st.columns(3)

with footer_col1:
    st.markdown("### 🎯 Your Goals")
    st.write("Personalized nutrition for better health")

with footer_col2:
    st.markdown("### 📊 Track Progress")
    st.write("Monitor your daily meal completion")

with footer_col3:
    st.markdown("### 🔄 Stay Flexible")
    st.write("Adjust your plan as needed")

st.markdown("---")
st.markdown(
    "🌟 Your journey to better health starts here! 🌟",
    unsafe_allow_html=True
)
```

```python
import streamlit as st
import pandas as pd
import os
from langchain.agents import Tool, AgentExecutor, create_react_agent
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
import google.generativeai as genai
import matplotlib.pyplot as plt
import re
from dotenv import load_dotenv

st.set_page_config(page_title="🥗 Personalized Meal Plan Generator", layout="wide")

load_dotenv()

# ---------- CONFIGURE API ----------
# Initialize the Search Tool
search = DuckDuckGoSearchRun()
tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Useful for searching information on the internet. Use this when you need to find current or factual information about food calories."
    )
]

# Initialize the language model
llm = ChatGroq(
    temperature=0.7,
    model="llama3-8b-8192",
    api_key=os.getenv("GROQ_API_KEY"),
)

api_key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.0-flash-001')

# ---------- LOAD DATASETS ----------
@st.cache_data
def load_meals(csv_path: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])
    return df

@st.cache_data
def load_nutrition_data(csv_path: str) -> pd.DataFrame:
    """Load the nutrition data CSV for calorie lookup"""
    try:
        df = pd.read_csv(csv_path)
        return df
    except FileNotFoundError:
        st.warning("nutrition_data.csv not found. Agent will be used for all unknown foods.")
        return pd.DataFrame()

meals_df = load_meals("meal.csv")
nutrition_df = load_nutrition_data("nutrition_data.csv")

# ---------- HELPER FUNCTION TO SAFELY CONVERT TO INT ----------
def safe_int_conversion(value):
    """Safely convert a value to integer, handling strings and None values"""
    if value is None:
        return None
    
    # If it's already a number
    if isinstance(value, (int, float)):
        return int(value)
    
    # If it's a string, try to extract number
    if isinstance(value, str):
        # Remove common text and extract numbers
        cleaned = re.sub(r'[^\d.]', '', str(value))
        if cleaned:
            try:
                return int(float(cleaned))
            except ValueError:
                # Try to find first number in the string
                match = re.search(r'(\d+)', str(value))
                if match:
                    return int(match.group(1))
    
    return None

# ---------- CALORIE LOOKUP FUNCTIONS ----------
def search_calories_in_nutrition_data(food_item: str) -> dict:
    """
    Search for food item in nutrition_data.csv
    Returns: dict with 'found', 'calories', 'portion' keys
    """
    if nutrition_df.empty:
        return {'found': False, 'calories': None, 'portion': None}
    
    # Normalize the food item name for better matching
    food_item_lower = food_item.lower().strip()
    
    # Search in the Food Item column (case-insensitive)
    matches = nutrition_df[nutrition_df['Food Item'].str.lower().str.contains(food_item_lower, na=False)]
    
    if not matches.empty:
        # Take the first match
        match = matches.iloc[0]
        # Safely convert calories to int
        calories = safe_int_conversion(match['Calories'])
        
        if calories is not None:
            return {
                'found': True,
                'calories': calories,
                'portion': match['Portion'],
                'food_name': match['Food Item']
            }
    
    return {'found': False, 'calories': None, 'portion': None}

def get_calories_from_agent(food_item: str) -> dict:
    """
    Use agent to search for calorie information online
    Returns: dict with 'found', 'calories', 'portion' keys
    """
    # Create the agent
    template = """Answer the following questions as best you can. You have access to the following tools:

{tools}

Use the following format:

Question: the input question you must answer
Thought: you should always think about what to do
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought: {agent_scratchpad}"""

    agent = create_react_agent(
        llm=llm,
        tools=tools,
        prompt=PromptTemplate.from_template(template)
    )

    agent_executor = AgentExecutor.from_agent_and_tools(
        agent=agent,
        tools=tools,
        verbose=False,
        memory=ConversationBufferMemory(memory_key="chat_history"),
        max_iterations=3
    )
    
    try:
        query = f"What are the calories in {food_item}? Provide just the calorie number per standard serving portion."
        response = agent_executor.invoke({"input": query})
        
        # Extract calories from response
        response_text = response.get('output', '')
        
        # Try to extract calorie number from response
        calorie_match = re.search(r'(\d+)\s*(?:cal|kcal|calories)', response_text.lower())
        
        if calorie_match:
            calories = int(calorie_match.group(1))
            return {
                'found': True,
                'calories': calories,
                'portion': 'standard serving',
                'source': 'agent_search'
            }
        
    except Exception as e:
        st.error(f"Agent search failed: {str(e)}")
    
    return {'found': False, 'calories': None, 'portion': None}

def get_food_calories(food_item: str) -> dict:
    """
    Main function to get calories for any food item
    First checks nutrition_data.csv, then uses agent as fallback
    """
    # Step 1: Check nutrition data CSV
    nutrition_result = search_calories_in_nutrition_data(food_item)
    
    if nutrition_result['found']:
        st.info(f"✅ Found '{food_item}' in nutrition database")
        return nutrition_result
    
    # Step 2: Use agent as fallback
    st.info(f"🔍 '{food_item}' not found in database. Searching online...")
    agent_result = get_calories_from_agent(food_item)
    
    if agent_result['found']:
        st.success(f"✅ Found calorie information for '{food_item}' online")
        return agent_result
    else:
        st.error(f"❌ Could not find calorie information for '{food_item}'")
        return {'found': False, 'calories': None, 'portion': None}

# ---------- BMI and BMR functions ----------
def classify_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)

def calculate_bmr(weight_kg, height_cm, age, gender):
    if gender.lower().startswith('m'):
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age + 5, 1)
    else:
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age - 161, 1)

# ---------- MEAL RECOMMENDATION ----------
def meal_recommendation(user_profile: dict, meals_df: pd.DataFrame) -> str:
    bmi_value = classify_bmi(user_profile['weight'], user_profile['height'])
    bmr = calculate_bmr(
        user_profile['weight'], user_profile['height'], user_profile['age'], user_profile['gender']
    )

    prompt = f"""
You are an expert dietitian assistant.

User Profile:
- Name: {user_profile['name']}
- Age: {user_profile['age']}
- Gender: {user_profile['gender']}
- Height: {user_profile['height']} cm
- Weight: {user_profile['weight']} kg
- BMI: {bmi_value}
- BMR: {bmr} kcal/day
- Health Goals: {', '.join(user_profile['health_goals'])}
- Food Allergies: {', '.join(user_profile['food_allergies'])}
- Dietary Pattern: {user_profile['diet']}
- Disease: {', '.join(user_profile['disease'])}
- Activity Level: {user_profile['activity_level']}
- Supplement Preferences: {', '.join(user_profile['supplement_preferences'])}

---

Task:
- Create a personalized meal plan with exactly 5 meals per day: Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, and Dinner.
- Choose **only** from this list of meals given below. Do not invent new meals.
- Each meal in the list has type (Breakfast, Snack, Lunch, Dinner), calories, protein, carbs, and fats.
- Consider user's BMI:
  * If BMI <18.5: Recommend meals above BMR (weight gain).
  * If BMI >=25: Recommend meals below BMR (weight loss).
  * If BMI 18.5–25: Recommend around BMR (maintenance).

---

📄 **Available Meals**:

{meals_df.to_string(index=False)}

---

✅ Format your answer like this:

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only pick from the provided list. Keep it personalized and explain briefly why you chose each.
"""

    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- MEAL MODIFICATION FUNCTION ----------
def handle_food_intake_modification(user_input: str, current_meal_plan: str) -> str:
    """
    Handle when user inputs food not in the suggested meal plan
    """
    # Extract food items mentioned by user
    food_items = extract_food_items_from_input(user_input)
    
    if not food_items:
        return "I couldn't identify specific food items in your request. Please mention the specific food you want to add or replace."
    
    # Get calorie information for each food item
    food_calorie_info = {}
    for food in food_items:
        calorie_info = get_food_calories(food)
        food_calorie_info[food] = calorie_info
    
    # Create modified meal plan prompt
    calorie_info_text = ""
    for food, info in food_calorie_info.items():
        if info['found']:
            calorie_info_text += f"- {food}: {info['calories']} calories per {info['portion']}\n"
    
    if not calorie_info_text:
        return "I couldn't find calorie information for the foods you mentioned. Please try with different food items."
    
    modification_prompt = f"""
You are an expert dietitian assistant.

Current meal plan:
{current_meal_plan}

User wants to modify the plan with: {user_input}

Available calorie information for user's food items:
{calorie_info_text}

Task:
- Modify the meal plan according to user's request
- Replace or add the mentioned foods appropriately
- Maintain similar total daily calories
- Keep the same format as the original plan

Provide the updated meal plan in the same format:
Breakfast: [Meal Name] (calories, protein, carbs, fats) - Justification.
Mid-Morning Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.
Lunch: [Meal Name] (calories, protein, carbs, fats) - Justification.
Afternoon Snack: [Meal Name] (calories, protein, carbs, fats) - Justification.
Dinner: [Meal Name] (calories, protein, carbs, fats) - Justification.
"""
    
    response = model.generate_content(modification_prompt)
    return response.text.strip()

def extract_food_items_from_input(user_input: str) -> list:
    """
    Extract food items from user input
    This is a simple implementation - you can make it more sophisticated
    """
    # Common food-related keywords that might indicate food items
    food_keywords = []
    
    # Split input into words and look for potential food items
    words = user_input.lower().split()
    
    # Simple approach: look for common food patterns
    potential_foods = []
    
    # Look for phrases like "add pizza", "replace with burger", etc.
    for i, word in enumerate(words):
        if word in ['add', 'replace', 'with', 'want', 'like', 'prefer']:
            # Check next few words for food items
            for j in range(i+1, min(i+4, len(words))):
                if j < len(words):
                    potential_foods.append(words[j])
    
    # Also check against nutrition database for any matches
    if not nutrition_df.empty:
        for word in words:
            matches = nutrition_df[nutrition_df['Food Item'].str.lower().str.contains(word, na=False)]
            if not matches.empty:
                potential_foods.append(word)
    
    # Remove duplicates and common non-food words
    non_food_words = {'with', 'and', 'the', 'a', 'an', 'some', 'more', 'less', 'want', 'like', 'prefer', 'add', 'replace'}
    food_items = list(set([food for food in potential_foods if food not in non_food_words and len(food) > 2]))
    
    return food_items

# ---------- MEAL CONFIGURATION BY USER (UPDATED) ----------
def meal_change_according_to_user(user_input: str, recommended_meal: str, meals_df: pd.DataFrame) -> str:
    prompt = f"""
You are an expert dietitian assistant.

Below is the meal plan previously recommended to the user:

{recommended_meal}

The user has requested the following change:
"{user_input}"

---

Your task:
- Identify which meal (Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, or Dinner) the user wants to change.
- From the given meals list below, select a new meal of the same type as the one to replace.
- The new meal should have similar calories (within ±30 kcal) and follow the user's original dietary needs.
- Do not select the same meal that was originally suggested.
- Keep all other meals unchanged.

📄 Available Meals:
{meals_df.to_string(index=False)}

---

✅ Output Format (same as before):

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only change the requested meal. Keep everything else exactly as it is.
"""

    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- HELPER: parse calories ----------
def parse_meal_calories(meal_text: str) -> dict:
    """Parse meal plan text to extract calories for each meal type"""
    calories = {}
    meal_types = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    for line in meal_text.splitlines():
        line = line.strip()
        if ':' in line and '(' in line:
            for meal_type in meal_types:
                if line.startswith(meal_type + ':'):
                    try:
                        # Extract calories from format like "(250 kcal" or "(250,"
                        match = re.search(r'\((\d+)', line)
                        if match:
                            calories[meal_type] = int(match.group(1))
                            break
                    except:
                        continue
    return calories

# ---------- ENHANCED CALORIE CALCULATION WITH QUANTITY ----------
def parse_food_with_quantity(user_input: str) -> dict:
    """
    Parse user input to extract food item and quantity
    Examples: "2 slices of pizza", "1 cup rice", "200g chicken"
    Returns: {'food': str, 'quantity': float, 'unit': str, 'raw_input': str}
    """
    user_input = user_input.strip().lower()
    
    # Common quantity patterns
    quantity_patterns = [
        r'(\d+(?:\.\d+)?)\s*(cups?|cup)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(slices?|slice)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(pieces?|piece)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*([kg|g|grams?|kilograms?])\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(bowls?|bowl)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(plates?|plate)\s+(?:of\s+)?(.+)',
        r'(\d+(?:\.\d+)?)\s*(.+)',  # Just number and food
    ]
    
    for pattern in quantity_patterns:
        match = re.search(pattern, user_input)
        if match:
            if len(match.groups()) == 3:
                quantity, unit, food = match.groups()
                return {
                    'food': food.strip(),
                    'quantity': float(quantity),
                    'unit': unit.strip(),
                    'raw_input': user_input
                }
            elif len(match.groups()) == 2:
                quantity, food = match.groups()
                return {
                    'food': food.strip(),
                    'quantity': float(quantity),
                    'unit': 'serving',
                    'raw_input': user_input
                }
    
    # If no quantity found, assume 1 serving
    return {
        'food': user_input,
        'quantity': 1.0,
        'unit': 'serving',
        'raw_input': user_input
    }

def calculate_adjusted_calories(food_info: dict, base_calories) -> int:
    """
    Calculate adjusted calories based on quantity and unit
    Fixed version with proper type checking
    """
    # Safely convert base_calories to int
    calories_int = safe_int_conversion(base_calories)
    
    if calories_int is None:
        st.error(f"Could not convert calories '{base_calories}' to a number")
        return 0
    
    quantity = food_info['quantity']
    unit = food_info['unit']
    
    # Define multipliers for different units
    unit_multipliers = {
        'serving': 1.0,
        'cup': 1.0,
        'slice': 0.8,  # Assuming slice is smaller than full serving
        'piece': 0.8,
        'bowl': 1.2,   # Bowl is usually larger
        'plate': 1.5,  # Plate is usually larger
        'g': 0.01,     # Per gram (assuming base is per 100g)
        'gram': 0.01,
        'grams': 0.01,
        'kg': 10.0,    # Per kg
        'kilogram': 10.0,
        'kilograms': 10.0,
    }
    
    multiplier = unit_multipliers.get(unit, 1.0)
    adjusted_calories = int(calories_int * quantity * multiplier)
    
    return adjusted_calories

def get_food_calories_with_quantity(user_input: str) -> dict:
    """
    Enhanced function to get calories considering quantity
    """
    # Parse food and quantity
    food_info = parse_food_with_quantity(user_input)
    
    # Get base calories for the food item
    calorie_result = get_food_calories(food_info['food'])
    
    if calorie_result['found']:
        # Calculate adjusted calories based on quantity
        adjusted_calories = calculate_adjusted_calories(food_info, calorie_result['calories'])
        
        return {
            'found': True,
            'food': food_info['food'],
            'quantity': food_info['quantity'],
            'unit': food_info['unit'],
            'base_calories': calorie_result['calories'],
            'adjusted_calories': adjusted_calories,
            'portion': calorie_result['portion'],
            'raw_input': food_info['raw_input']
        }
    
    return {'found': False, 'raw_input': user_input}

# ---------- MEAL ADJUSTMENT BASED ON CALORIE DIFFERENCE ----------
def adjust_next_meals(consumed_food_info: dict, current_meal_plan: str, meal_eaten: str) -> str:
    """
    Adjust the next meals based on calorie difference from what was consumed vs planned
    """
    # Parse current meal plan to get planned calories
    planned_calories = parse_meal_calories(current_meal_plan)
    
    if meal_eaten not in planned_calories:
        return "Could not identify which meal to adjust. Please specify (Breakfast, Lunch, Dinner, etc.)"
    
    # Calculate calorie difference
    planned_meal_calories = planned_calories[meal_eaten]
    consumed_calories = consumed_food_info['adjusted_calories']
    calorie_difference = consumed_calories - planned_meal_calories
    
    # Get meal order to determine next meals
    meal_order = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    
    try:
        current_meal_index = meal_order.index(meal_eaten)
        remaining_meals = meal_order[current_meal_index + 1:]
    except ValueError:
        remaining_meals = []
    
    if not remaining_meals:
        return f"""
✅ **Meal Update Complete**

You consumed: **{consumed_food_info['food'].title()}** ({consumed_food_info['quantity']} {consumed_food_info['unit']})
- Calories: {consumed_calories} kcal
- Planned: {planned_meal_calories} kcal
- Difference: {'+' if calorie_difference > 0 else ''}{calorie_difference} kcal

Since this was your last meal of the day, no further adjustments needed. 
{('Consider lighter meals tomorrow if you exceeded your target.' if calorie_difference > 0 else 'Great job staying within your calorie goals!')}
        """
    
    # Create adjustment prompt
    adjustment_prompt = f"""
You are an expert dietitian assistant.

Current meal plan:
{current_meal_plan}

**Meal Adjustment Needed:**
- User consumed: {consumed_food_info['food']} ({consumed_food_info['quantity']} {consumed_food_info['unit']}) = {consumed_calories} kcal
- Originally planned for {meal_eaten}: {planned_meal_calories} kcal  
- Calorie difference: {calorie_difference} kcal {'(excess)' if calorie_difference > 0 else '(deficit)'}

**Remaining meals to adjust:** {', '.join(remaining_meals)}

**Task:**
1. Keep the consumed meal as: "{meal_eaten}: {consumed_food_info['food'].title()} ({consumed_food_info['quantity']} {consumed_food_info['unit']}) ({consumed_calories} kcal) - User consumed this instead"
2. Adjust the remaining meals ({', '.join(remaining_meals)}) to compensate for the {calorie_difference} kcal difference
3. If excess calories: Choose lower-calorie alternatives from the meal database
4. If deficit calories: Choose slightly higher-calorie alternatives or add healthy additions
5. Distribute the adjustment across remaining meals
6. Select only from the provided meals database

Available Meals Database:
{meals_df.to_string(index=False)}

✅ Format your answer like this:

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only pick from the provided list. Keep it personalized and explain briefly why you chose each.

"""
    
    response = model.generate_content(adjustment_prompt)
    return response.text.strip()

# ---------- UPDATED UI SECTION FOR ADJUSTMENT ----------
# Replace the existing adjustment form in col3 with this enhanced version:

def render_enhanced_adjustment_section():
    """Enhanced adjustment section UI"""
    st.subheader("🍽️ Ate Something Different?")
    st.write("Tell me what you ate and I'll adjust your remaining meals!")
    
    with st.form("enhanced_adjustment_form"):
        # Meal selection
        meal_eaten = st.selectbox(
            "Which meal did you just have?",
            options=["Breakfast", "Mid-Morning Snack", "Lunch", "Afternoon Snack", "Dinner"],
            help="Select the meal you want to replace"
        )
        
        # Food input with quantity
        food_consumed = st.text_input(
            "What did you eat? (Include quantity)",
            placeholder="e.g., 2 slices pizza, 1 cup rice, 200g chicken",
            help="Be specific about quantity for accurate calorie calculation"
        )
        
        # Examples
        with st.expander("💡 Quantity Examples"):
            st.write("""
            **Valid formats:**
            - 2 slices of pizza
            - 1 cup rice  
            - 200g chicken breast
            - 1 bowl pasta
            - 3 pieces samosa
            - 1 plate biryani
            """)
        
        adjust_submitted = st.form_submit_button("🔄 Adjust My Meal Plan", type="primary")
        
        if adjust_submitted and food_consumed:
            with st.spinner("🔄 Calculating calories and adjusting your plan..."):
                # Get food info with quantity
                food_info = get_food_calories_with_quantity(food_consumed)
                
                if food_info['found']:
                    # Show what was consumed
                    st.success(f"✅ **Food Logged:** {food_info['food'].title()}")
                    
                    col1, col2 = st.columns(2)
                    with col1:
                        st.metric("Quantity", f"{food_info['quantity']} {food_info['unit']}")
                    with col2:
                        st.metric("Total Calories", f"{food_info['adjusted_calories']} kcal")
                    
                    # Adjust the meal plan
                    adjusted_plan = adjust_next_meals(food_info, st.session_state['meal_plan'], meal_eaten)
                    st.session_state['meal_plan'] = adjusted_plan
                    
                    st.success("✅ Your meal plan has been adjusted!")
                    st.rerun()
                else:
                    st.error("Could not find calorie information for this food item. Please try with a different food name.")

# ---------- STREAMLIT UI (Rest remains same) ----------
st.title("🥗 Personalized Meal Plan Generator")
st.markdown("*Create customized meal plans based on your health goals and dietary preferences*")
st.divider()

# Create layout columns
col1, col2, col3 = st.columns([4, 2, 2], gap="large")

with col1:
    st.header("👤 Tell us about yourself")
    
    # User Input Form
    with st.form("user_info_form", clear_on_submit=False):
        # Personal Information Section
        st.subheader("📋 Personal Details")
        
        # Two column layout for personal info
        pers_col1, pers_col2 = st.columns(2)
        
        with pers_col1:
            name = st.text_input("Name", placeholder="Enter your name")
            age = st.number_input("Age", min_value=5, max_value=120, step=1, value=25)
            height = st.number_input("Height (cm)", min_value=100.0, max_value=250.0, step=1.0, value=170.0)
        
        with pers_col2:
            food_type = st.selectbox(
                "Preferred Cuisine",
                options=meals_df['Cuisine'].unique() if 'Cuisine' in meals_df.columns else ['Indian', 'Continental', 'Asian', 'Mediterranean']
            )
            gender = st.selectbox("Gender", ["Male", "Female", "Other"])
            weight = st.number_input("Weight (kg)", min_value=20.0, max_value=300.0, step=1.0, value=70.0)
        
        st.divider()
        
        # Health Goals Section
        st.subheader("🎯 Health & Fitness Goals")
        health_goals = st.multiselect(
            "What are your primary health goals? (Select all that apply)",
            options=[
                "Weight management", "Energy boost", "Hormonal balance", "Bone & joint health",
                "Muscle gain", "Skin, hair & nails", "Gut health", "Immune support",
                "Fertility & reproductive health", "Menstrual wellness / Menopause support", "Other"
            ],
            help="Choose your main health objectives to get personalized recommendations"
        )
        
        st.divider()
        
        # Dietary Preferences Section
        st.subheader("🍽️ Dietary Information")
        diet_col1, diet_col2 = st.columns(2)
        
        with diet_col1:
            diet = st.selectbox(
                "Dietary Pattern",
                options=["Omnivore", "Vegetarian", "Vegan", "Pescatarian", "Low-carb/Keto", "Gluten-free", "Jain/Sattvic", "Other"]
            )
            
            activity_level = st.selectbox(
                "Activity Level",
                options=[
                    "Little movement",
                    "Walking, stretching everyday",
                    "Exercise 2–3 times a week",
                    "Daily training"
                ]
            )
        
        with diet_col2:
            food_allergies = st.multiselect(
                "Food Allergies & Intolerances",
                options=[
                    "Dairy (lactose or casein)", "Gluten (celiac or intolerance)", "Soy", "Peanuts",
                    "Tree nuts", "Eggs", "Shellfish", "Fish", "Corn",
                    "Nightshades (tomatoes, peppers)", "Artificial sweeteners", "Caffeine", "Alcohol"
                ]
            )
        
        st.divider()
        
        # Health Conditions Section
        st.subheader("🏥 Health Conditions")
        disease = st.multiselect(
            "Do you currently manage any of these conditions?",
            options=[
                "Diabetes / Insulin resistance", "PCOS/PCOD", "High blood pressure / cholesterol",
                "Thyroid imbalance (Hypo/Hyper)", "Hormonal Imbalance", "Digestive disorders (IBS, IBD, acidity)",
                "Menstrual issues or menopause symptoms", "Autoimmune disease"
            ]
        )
        
        st.divider()
        
        # Supplement Preferences Section
        st.subheader("💊 Supplement Preferences (Optional)")
        supplement_preferences = st.multiselect(
            "What type of supplements do you prefer?",
            options=[
                "Herbal options", "Clinical-grade with scientific backing", "Vegan options",
                "Gluten-free options", "Gelatin-free options", "Soy-free options", "Corn-free options",
                "Cruelty-free options"
            ]
        )
        
        # Submit button
        st.markdown("")
        submitted = st.form_submit_button(
            "🚀 Generate My Meal Plan",
            type="primary",
            use_container_width=True
        )

# Generate Meal Plan
if submitted:
    user_profile = {
        "name": name,
        "age": age,
        "gender": gender,
        "height": height,
        "weight": weight,
        "health_goals": health_goals,
        "food_allergies": food_allergies,
        "diet": diet,
        "disease": disease,
        "activity_level": activity_level,
        "supplement_preferences": supplement_preferences
    }

    with st.spinner("🔄 Creating your personalized meal plan..."):
        meal_plan_text = meal_recommendation(user_profile, meals_df)
        st.session_state['meal_plan'] = meal_plan_text

    # Initialize meal tracking
    if 'meal_completion' not in st.session_state:
        st.session_state['meal_completion'] = {}

    st.success("✅ Your meal plan is ready!")

# Display Meal Plan
if 'meal_plan' in st.session_state:
    st.divider()
    st.header("🍽️ Your Personalized Meal Plan")
    st.markdown(st.session_state['meal_plan'])

    # Meal Modification Section
    st.divider()
    st.header("🔄 Modify Your Plan")
    
    with st.form("modification_form"):
        st.write("Need changes to your meal plan? Let me know!")
        st.info("💡 You can ask for foods not in our meal database - I'll search for their calorie information online!")
        
        user_input = st.text_area(
            "Describe the changes you'd like:",
            placeholder="e.g., 'Replace chicken with tofu', 'Add pizza to lunch', 'I want burger instead'"
        )

        change_submitted = st.form_submit_button("✨ Apply Changes", type="primary")

        if change_submitted and user_input:
            with st.spinner("🔄 Updating your meal plan..."):
                updated_meal = meal_change_according_to_user(user_input, st.session_state['meal_plan'], meals_df)
                st.session_state['meal_plan'] = updated_meal
                st.success("✅ Your meal plan has been updated!")
                st.rerun()

# Progress Tracking Column
with col2:
    st.header("📊 Daily Progress")
    if 'meal_plan' in st.session_state:
        calories_map = parse_meal_calories(st.session_state['meal_plan'])
        
        if calories_map:
            total_required = sum(calories_map.values())
            
            # Initialize tracking
            if 'meal_completion' not in st.session_state:
                st.session_state['meal_completion'] = {meal: False for meal in calories_map.keys()}

            st.subheader("✅ Track Your Meals")
            st.write("Check off meals as you complete them:")

            consumed_calories = 0

            # Meal checkboxes
            for meal_type, cal in calories_map.items():
                completed = st.checkbox(
                    f"{meal_type} ({cal} kcal)",
                    key=f"meal_{meal_type}",
                    value=st.session_state['meal_completion'].get(meal_type, False)
                )

                st.session_state['meal_completion'][meal_type] = completed
                if completed:
                    consumed_calories += cal

            # Progress metrics
            remaining_calories = max(total_required - consumed_calories, 0)
            progress_percentage = (consumed_calories / total_required) * 100 if total_required > 0 else 0

            st.divider()

            # Display metrics using Streamlit's metric component
            metric_col1, metric_col2 = st.columns(2)

            with metric_col1:
                st.metric("Consumed", f"{consumed_calories} kcal")

            with metric_col2:
                st.metric("Remaining", f"{remaining_calories} kcal")

            st.metric("Progress", f"{progress_percentage:.1f}%")

            # Progress bar
            st.progress(progress_percentage / 100)

            # Status messages
            if consumed_calories >= total_required:
                st.success("🎉 Daily goal completed!")
            elif consumed_calories > 0:
                st.info(f"📈 {progress_percentage:.1f}% complete")
            else:
                st.warning("🍽️ Start tracking your meals!")

            # Visual chart
            if consumed_calories > 0 or remaining_calories > 0:
                st.subheader("📈 Visual Progress")
                fig, ax = plt.subplots(figsize=(8, 6))
                labels = []
                sizes = []
                colors = []

                if consumed_calories > 0:
                    labels.append(f'Consumed\n{consumed_calories} kcal')
                    sizes.append(consumed_calories)
                    colors.append('#00C851')

                if remaining_calories > 0:
                    labels.append(f'Remaining\n{remaining_calories} kcal')
                    sizes.append(remaining_calories)
                    colors.append('#ffbb33')

                wedges, texts, autotexts = ax.pie(
                    sizes,
                    labels=labels,
                    autopct='%1.1f%%',
                    startangle=90,
                    colors=colors
                )

                ax.set_title('Daily Calorie Progress', fontsize=14, fontweight='bold')
                ax.axis('equal')
                st.pyplot(fig)
                plt.close()

        else:
            st.error("Unable to parse meal plan calories")
    else:
        st.info("🍽️ Generate a meal plan to start tracking your progress!")

# Adjustment Helper Column
with col3:
    st.header("🔧 Smart Adjustments")
    
    if 'meal_plan' in st.session_state:
        render_enhanced_adjustment_section()
        
        st.divider()

        # Quick tips
        st.subheader("💡 Quick Tips")
        tips = [
            "🥤 Drink water before meals",
            "🥗 Fill half your plate with vegetables",
            "🚶‍♀️ Take a 5-minute walk after eating",
            "⏰ Eat mindfully and slowly",
            "😴 Get adequate sleep for better metabolism"
        ]

        for tip in tips:
            st.write(tip)
    else:
        st.info("🍽️ Generate a meal plan first to access adjustment tools!")

        # General wellness tips while waiting
        st.subheader("🌟 Wellness Tips")
        st.write("While you create your meal plan:")
        
        general_tips = [
            "💧 Aim for 8 glasses of water daily",
            "🥬 Include colorful vegetables",
            "🥜 Add healthy fats like nuts and seeds",
            "🏃‍♀️ Move your body daily",
            "😌 Practice mindful eating"
        ]

        for tip in general_tips:
            st.write(tip)

# Footer
st.divider()
st.markdown("---")

# Summary footer
footer_col1, footer_col2, footer_col3 = st.columns(3)

with footer_col1:
    st.markdown("### 🎯 Your Goals")
    st.write("Personalized nutrition for better health")

with footer_col2:
    st.markdown("### 📊 Track Progress")
    st.write("Monitor your daily meal completion")

with footer_col3:
    st.markdown("### 🔄 Stay Flexible")
    st.write("Adjust your plan as needed")

st.markdown("---")
st.markdown(
    "🌟 Your journey to better health starts here! 🌟",
    unsafe_allow_html=True
)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
import pandas as pd
import google.generativeai as genai
import re
from typing import List, Dict

app = FastAPI()

# ---------- CONFIGURE API ----------
api_key = "AIzaSyBMdmH6lJdWLtqQ4YBB8bkqXaGukQm_47s"
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.0-flash-001')

# ---------- LOAD MEAL CSV ----------
meals_df = pd.read_csv("meal.csv")
if 'Unnamed: 0' in meals_df.columns:
    meals_df.drop(columns=['Unnamed: 0'], inplace=True)

# ---------- BMI and BMR functions ----------
def classify_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)

def calculate_bmr(weight_kg, height_cm, age, gender):
    if gender.lower().startswith('m'):
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age + 5, 1)
    else:
        return round(10 * weight_kg + 6.25 * height_cm - 5 * age - 161, 1)

# ---------- MODELS ----------
class UserProfile(BaseModel):
    name: str
    age: int
    gender: str
    height: float
    weight: float
    health_goals: List[str]
    food_allergies: List[str]
    diet: str
    disease: List[str]
    activity_level: str
    supplement_preferences: List[str]

class MealChangeRequest(BaseModel):
    user_input: str
    recommended_meal: str

class MealCompletionRequest(BaseModel):
    meal_text: str
    completed_meals: List[str]

# ---------- MEAL RECOMMENDATION ----------
def meal_recommendation(user_profile: dict, meals_df: pd.DataFrame) -> str:
    bmi_value = classify_bmi(user_profile['weight'], user_profile['height'])
    bmr = calculate_bmr(
        user_profile['weight'], user_profile['height'], user_profile['age'], user_profile['gender']
    )
    prompt = f"""
You are an expert dietitian assistant.

User Profile:
- Name: {user_profile['name']}
- Age: {user_profile['age']}
- Gender: {user_profile['gender']}
- Height: {user_profile['height']} cm
- Weight: {user_profile['weight']} kg
- BMI: {bmi_value}
- BMR: {bmr} kcal/day
- Health Goals: {', '.join(user_profile['health_goals'])}
- Food Allergies: {', '.join(user_profile['food_allergies'])}
- Dietary Pattern: {user_profile['diet']}
- Disease: {', '.join(user_profile['disease'])}
- Activity Level: {user_profile['activity_level']}
- Supplement Preferences: {', '.join(user_profile['supplement_preferences'])}

---

Task:
- Create a personalized meal plan with exactly 5 meals per day: Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, and Dinner.
- Choose **only** from this list of meals given below. Do not invent new meals.
- Each meal in the list has type (Breakfast, Snack, Lunch, Dinner), calories, protein, carbs, and fats.
- Consider user's BMI:
  * If BMI <18.5: Recommend meals above BMR (weight gain).
  * If BMI >=25: Recommend meals below BMR (weight loss).
  * If BMI 18.5–25: Recommend around BMR (maintenance).

---

📄 **Available Meals**:

{meals_df.to_string(index=False)}

---

✅ Format your answer like this:

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only pick from the provided list. Keep it personalized and explain briefly why you chose each.
"""
    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- MEAL CONFIGURATION BY USER ----------
def meal_change_according_to_user(user_input: str, recommended_meal: str, meals_df: pd.DataFrame) -> str:
    prompt = f"""
You are an expert dietitian assistant.

Below is the meal plan previously recommended to the user:

{recommended_meal}

The user has requested the following change:
"{user_input}"

---

Your task:
- Identify which meal (Breakfast, Mid-Morning Snack, Lunch, Afternoon Snack, or Dinner) the user wants to change.
- From the given meals list below, select a new meal of the same type as the one to replace.
- The new meal should have similar calories (within ±30 kcal) and follow the user's original dietary needs.
- Do not select the same meal that was originally suggested.
- Keep all other meals unchanged.

📄 Available Meals:
{meals_df.to_string(index=False)}

---

✅ Output Format (same as before):

Breakfast: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Mid-Morning Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Lunch: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Afternoon Snack: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Dinner: <Meal name> (calories, protein, carbs, fats) - One-line justification.

Only change the requested meal. Keep everything else exactly as it is.
"""
    response = model.generate_content(prompt)
    return response.text.strip()

# ---------- PARSE CALORIES ----------
def parse_meal_calories(meal_text: str) -> Dict[str, int]:
    calories = {}
    meal_types = ['Breakfast', 'Mid-Morning Snack', 'Lunch', 'Afternoon Snack', 'Dinner']
    for line in meal_text.splitlines():
        for meal_type in meal_types:
            if line.startswith(meal_type + ':'):
                match = re.search(r'\((\d+)', line)
                if match:
                    calories[meal_type] = int(match.group(1))
    return calories

# ---------- API ENDPOINTS ----------
@app.post("/generate-meal")
def generate_meal(user: UserProfile):
    meal_plan = meal_recommendation(user.dict(), meals_df)
    return {"meal_plan": meal_plan}

@app.post("/change-meal")
def change_meal(change_req: MealChangeRequest):
    updated_meal = meal_change_according_to_user(change_req.user_input, change_req.recommended_meal, meals_df)
    return {"updated_meal": updated_meal}

@app.post("/meal-progress")
def meal_progress(data: MealCompletionRequest):
    calories_map = parse_meal_calories(data.meal_text)
    total_required = sum(calories_map.values())
    consumed = sum(calories_map[m] for m in data.completed_meals if m in calories_map)
    remaining = max(total_required - consumed, 0)
    percentage = (consumed / total_required) * 100 if total_required else 0

    return {
        "total_required": total_required,
        "consumed": consumed,
        "remaining": remaining,
        "progress_percent": round(percentage, 1),
        "meal_breakdown": calories_map
    }
```
